```python
# Drop the columns in "Dress Sales" which have more than 40% of missing values.

fun = inp1.isnull().sum()
per = fun[fun > 40].index
inp1.drop(per, axis=1, inplace=True)
inp1.info()

#df = inp1.loc[:, per]  (failed approch, kept for future refrance)
#df
```

```python
#Grouping columns in new columns
inp1['Summer'] = inp1['29-08-2013'] + inp1['31-08-2013'] + inp1['10-06-2013']
inp1['Autumn'] = inp1['09-10-2013'] + inp1['14-09-2013']+ inp1['16-09-2013']+ inp1['18-09-2013']+ inp1['20-09-2013']+ inp1['22-09-2013']+ inp1['24-09-2013']+ inp1['28-09-2013']
inp1['Winter'] = inp1['10-12-2013'] + inp1['09-02-2013'] + inp1['09-12-2013']
inp1['Spring'] = inp1['09-04-2013']

```

```python
#merge : Left Join
inp0 = pd.merge(left=inp0, right=inp1, how='left', left_on='Dress_ID', right_on='Dress_ID')
inp0.head()
```

```python
#drop unwanted columns, using range of columns as indexer
inp0.drop(inp0.loc[:, '29-08-2013':'10-12-2013'].columns, axis=1, inplace=True)
inp0.isnull().sum()
```

```python
#handling missing values by dropping acording to their low volumne
less_miss = ['Price','Season','NeckLine','SleeveLength','Winter','Autumn']
inp0.dropna(subset = less_miss, inplace=True)

```

```python
#handling missing values by replacing them acording to their high volumne
inp0['Material'] = inp0['Material'].replace(np.nan, 'Missing')
inp0.FabricType = inp0['FabricType'].replace(np.nan, 'Missing')
inp0.Decoration = inp0['Decoration'].replace(np.nan, 'Missing')
inp0['Pattern Type'] = inp0['Pattern Type'].replace(np.nan, 'Missing')
```
